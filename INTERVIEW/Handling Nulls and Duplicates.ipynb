```python
from pyspark.sql.functions import *
from pyspark.sql import *
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```python
data = [(1,"A",12),
        (2,"B",None),
        (3,None,None),
        (None,"D",11)]

mySchema = StructType([StructField("ID",IntegerType(),True),
                       StructField("Name",StringType(),True),
                       StructField("Age",IntegerType(),True),])

df = spark.createDataFrame(data,mySchema)
df.show()
```

```python
df.printSchema()
```

### Filling 'Null' from value = 0 (Zero)
- Only integer type Column Value can be raplaced with 0, String Type Column will still be null

```python
df.na.fill(value=0).show()
df.na.fill(0).show()
df.fillna(0)
df.show()
```

### Replacing String Type Column Value NULL with Empty String

```python
df.fillna("").show()
```

### Replacing String Type NULL value with Empty String and Integer Type String Value with 0 (Zero)

```python
df.fillna(0).fillna("").show()
```

### Replacing The Null Value Of only Age Column with 0 (zero)

```python
df.fillna(0,['Age']).show()
```

### Replacing The Vaule NULL Of More 1 Column (ONLY SELECTIVE COLUMNS)

```python
df.fillna(0,['Age','ID']).show()
```
