Q1 : Give Top 2 Salary Of Each Department

- We Are not Using Rank and Dense Rank , becoz when we filter top 2 salary at that time if top salary having more the 1 people all top will get same rank, which will lead more than 2 record coming out from that condition but we want ONLY TOP 2 reocrd from each department

```python
from pyspark.sql.window import Window

window_spec = Window.partitionBy("department").orderBy(desc("salary"))

top2_salary = df.withColumn("Rank",row_number().over(window_spec)).filter(col("Rank") <= 2)

top2_salary.show()
```
