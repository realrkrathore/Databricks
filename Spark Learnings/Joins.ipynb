```python
from pyspark.sql import *
from pyspark.sql.functions import *
```

```python

# Employee DataFrame
emp_data = [
    (1, "Rajat", 10),
    (2, "Amit", 20),
    (3, "Neha", 30),
    (4, "Priya", None),
    (5, "Vikas", 50)
]

emp_cols = ["emp_id", "emp_name", "dept_id"]

df_emp = spark.createDataFrame(emp_data, emp_cols)

# Department DataFrame
dept_data = [
    (10, "HR"),
    (20, "IT"),
    (30, "Finance"),
    (40, "Marketing")
]

dept_cols = ["dept_id", "dept_name"]

df_dept = spark.createDataFrame(dept_data, dept_cols)

df_emp.show()
df_dept.show()
```

#### Inner JOIN

```python
df_emp.join(df_dept,on = df_emp["dept_id"] == df_dept["dept_id"], how = "inner").show()

#OR

df_emp.join(df_dept,df_emp["dept_id"] == df_dept["dept_id"],"inner").show()
```

### Left Join

```python
df_emp.join(df_dept, on = df_emp["dept_id"] == df_dept["dept_id"], how = "left").show()

# OR

df_emp.join(df_dept, df_emp["dept_id"] == df_dept["dept_id"],"left").show()
```

-  Above We can see "dept_id" column is Coming 2 times we can Drop it.
- We can Also select The Selective Columns Using "Select" Statements

```python
# Droping 1 of the Dept_id Column
df_emp.join(df_dept, df_emp["dept_id"] == df_dept["dept_id"],"left").drop(df_dept["dept_id"]).show()

# Selecting only emp_id , emp_name, dept_name

df_emp.join(df_dept, df_emp["dept_id"] == df_dept["dept_id"],"left").select(df_emp["emp_id"],df_emp["emp_name"],df_dept["dept_name"]).show()
```

### Right JOIN

```python
df_emp.join(df_dept, on = df_emp["dept_id"] == df_dept["dept_id"], how = "right").show()

# OR

df_emp.join(df_dept, df_emp["dept_id"] == df_dept["dept_id"],  "right").show()
```

### Full Outer JOIN

```python
df_emp.join(df_dept, on = df_emp["dept_id"] == df_dept["dept_id"], how = "full_outer").show()

# OR

df_emp.join(df_dept, df_emp["dept_id"] == df_dept["dept_id"],  "full_outer").show()

# OR

df_emp.join(df_dept, on = df_emp["dept_id"] == df_dept["dept_id"], how = "outer").show()

# OR

df_emp.join(df_dept, df_emp["dept_id"] == df_dept["dept_id"],  "outer").show()
```

#### LEFT ANTI
- Shows ONLY NON - MATCHING records of LEFT TABLE

```python
df_emp.join(df_dept, on = df_emp["dept_id"] == df_dept["dept_id"], how = 'left_anti').show()

# OR

df_emp.join(df_dept, df_emp["dept_id"] == df_dept["dept_id"], 'left_anti').show()
```

#### LEFT-SEMI JOIN
- Shows Only Records of Left table , which have matcching key in Right table

```python
df_emp.join(df_dept, on = df_emp["dept_id"] == df_dept["dept_id"], how = "left_semi").show()

# OR

df_emp.join(df_dept, df_emp["dept_id"] == df_dept["dept_id"],  "left_semi").show()
```

#### Self JOIN 
- We need to use Alias

```python
df_emp_alias1 = df_emp.alias("e1")
df_emp_alias2 = df_emp.alias("e2") 

df_emp_alias1.join(df_emp_alias2, df_emp_alias1["emp_id"] == df_emp_alias2["emp_id"], "inner").show()
```
